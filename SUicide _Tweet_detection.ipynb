```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
df = pd.read_csv('Suicide.csv')
df
```

```python
df.groupby('Suicide').describe()
```

```python
df["Suicide"].value_counts()
```

```python
df.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df.Suicide = le.fit_transform(df.Suicide )
```

```python
df
```

```python
df=df.dropna()
df_Suicide = df[df['Suicide']==1]
df_Suicide.shape
```

```python
df_NS = df[df['Suicide']==0]
df_NS.shape
```

```python
df_NS_downsampled = df_NS.sample(df_Suicide.shape[0])
df_NS_downsampled.shape
```

```python
df_balanced = pd.concat([df_NS_downsampled, df_Suicide])
df_balanced.shape
```

```python
df_balanced['Suicide'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_balanced['Tweet'],df_balanced['Suicide'], stratify=df_balanced['Suicide'])
```

```python
X_train.head(4)
```

```python
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

```python
import tensorflow as tf
```

```python
# Bert layers
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)

# Neural network layers
l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs['pooled_output'])
l = tf.keras.layers.Dense(1, activation='sigmoid', name="output")(l)

# Use inputs and outputs to construct a final model
model = tf.keras.Model(inputs=[text_input], outputs = [l])
```

```python
model.summary()
```

```python
len(X_train)
```

```python
METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall')
]

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=METRICS)
```

```python
import numpy as np


y_train = np.asarray(y_train).astype(np.int32)
```

```python
len(X_train)
```

```python
df["Tweet"].dtypes
```

```python
model.fit(X_train, y_train, epochs=15)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted = y_predicted.flatten()
```

```python
import numpy as np

y_predicted = np.where(y_predicted > 0.5, 1, 0)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
